## Read and show image use cv2

```python
import cv2

# Đọc hình ảnh
img = cv2.imread("flowers.jpg")
cv2.imshow("Original", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Read and show image use matplotlib

```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh
img = cv2.imread("flowers.jpg")

# Hiển thị ảnh
plt.imshow(img)
plt.axis("off")
plt.show()
```

Because they use BGR color, so we have to change to RGB Color

```python
import cv2
import matplotlib.pyplot as plt

# Đọc hình ảnh
img = cv2.imread("flowers.jpg")

# Đổi thứ tự các kênh màu
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Hiển thị ảnh
plt.imshow(img)
plt.axis("off")
plt.show()
```
